```python
import pandas as pd
import json

seasons = range(2022, 2023)

gplus_data = pd.DataFrame()
for yr in seasons:
    print(f"Grabbing G+ for {yr}")
    url = f"https://app.americansocceranalysis.com/api/v1/mls/players/goals-added?season_name={yr}&split_by_teams=true&split_by_seasons=true"
    tmp = pd.read_json(url)
    tmp['season'] = yr
    gplus_data = gplus_data.append(tmp, ignore_index=True)
    
    gk_url = f"https://app.americansocceranalysis.com/api/v1/mls/goalkeepers/goals-added?season_name={yr}&split_by_teams=true&split_by_seasons=true"
    gk_tmp = pd.read_json(gk_url)
    gk_tmp['season'] = yr
    gk_tmp['general_position'] = 'GK'
    gplus_data = gplus_data.append(gk_tmp, ignore_index=True)
    
gplus_data
```

```python
salaries = pd.DataFrame()

for yr in seasons:
    print(f"Grabbing $ for {yr}")
    url = f"https://app.americansocceranalysis.com/api/v1/mls/players/salaries?season_name={yr}"
    tmp = pd.read_json(url)
    tmp['season'] = yr
    salaries = salaries.append(tmp, ignore_index=True)
    
salaries
```

```python
salaries = salaries[['season_name','player_id', 'team_id','guaranteed_compensation']]
salaries
```

```python
player_list = salaries[salaries.player_id.notna() == True].player_id.unique().tolist()
player_list
```

```python
import numpy as np

player_data = pd.DataFrame()
chunks = np.array_split(player_list, 50)
for i in range(0, len(chunks)):
    chunk = chunks[i]
    id_list = ','.join(chunk)
    print(f"Grabbing ids for chunk {i} of {len(chunks)}")
    url = f"https://app.americansocceranalysis.com/api/v1/mls/players?player_id={id_list}"
    tmp = pd.read_json(url)
    player_data = player_data.append(tmp, ignore_index=True)
    
player_data
```

```python
json_expl_txt = json.loads(gplus_data.explode('data').to_json(orient="records"))
gplus_expl_flat = pd.json_normalize(json_expl_txt)
gplus_expl_flat.head()
```

```python
# player_data.birth_date = player_data.birth_date.astype(str)
player_data['birth_datetime'] = pd.to_datetime(player_data.birth_date)
player_data['birth_year'] = player_data.birth_datetime.dt.year
# player_data['league_age'] = pd.Timestamp.today().year - player_data.birth_year
player_data
```

```python
player_salary = pd.merge(salaries, player_data, left_on="player_id", right_on="player_id")
player_salary = player_salary[["season_name_x","team_id","player_id","player_name","birth_year","guaranteed_compensation","primary_general_position"]]
player_salary['league_age'] = player_salary.season_name_x.astype(int) - player_salary.birth_year
player_salary
```

```python
grouped_gplus = gplus_expl_flat.groupby(['season_name','team_id','player_id','general_position']).agg({
    'data.goals_added_raw': ['sum'], 
    'minutes_played' : ['mean']
}).reset_index()
grouped_gplus.columns = grouped_gplus.columns.droplevel(level=1)
grouped_gplus['total'] = grouped_gplus['data.goals_added_raw']
grouped_gplus['p96'] = grouped_gplus['data.goals_added_raw'] * 96 / grouped_gplus["minutes_played"]
grouped_gplus = grouped_gplus[['season_name', 'team_id', 'player_id','general_position', 'minutes_played','total', 'p96']]

grouped_gplus
```

```python
salaried_gplus = pd.merge(grouped_gplus, player_salary, left_on=["season_name", "team_id","player_id"],right_on=["season_name_x", "team_id","player_id"])
salaried_gplus
```

```python
salaried_gplus['DP'] = salaried_gplus.player_id.isin(DPs)
salaried_gplus['YDP'] = (salaried_gplus.DP) & (salaried_gplus.league_age <= 23)
salaried_gplus['OldDP'] = (salaried_gplus.DP) & (salaried_gplus.league_age > 24)
salaried_gplus['TAM'] = (salaried_gplus.guaranteed_compensation >= 612500) & (salaried_gplus.guaranteed_compensation < 1612500)
salaried_gplus['U22'] = salaried_gplus.player_id.isin(U22s)
salaried_gplus['roster_desig'] = np.select(
[
    salaried_gplus['YDP'] == True,
    salaried_gplus['OldDP'] == True,
    salaried_gplus['U22'] == True,
    salaried_gplus['TAM'] == True
],
[
    "YDP",
    "DP",
    "U22",
    "TAM"
],default = "Normal")
salaried_gplus
```

```python
team_roster_gplus = salaried_gplus.groupby(['season_name','team_id','roster_desig']).agg({
    'roster_desig' : ['count'],
    'p96': ['mean'], 
    'minutes_played' : ['sum']
}).reset_index()
team_roster_gplus.columns = team_roster_gplus.columns.droplevel(level=1)
team_roster_gplus.columns = ['season', 'team_id', 'roster_desig', 'count', 'g+/96','total_min']
team_roster_gplus
```

```python

# team_roster_gplus
counts = team_roster_gplus.reset_index().groupby(['season','team_id','roster_desig'])['count'].aggregate('first').unstack()
counts
```

```python
counts = counts.reset_index().rename_axis(None, axis=1)
counts['category'] = np.select(
[
    (counts['DP'] == 2) & (counts['YDP'] == 1) & (counts['U22'] >= 2),
    (counts['DP'] == 2) & (counts['YDP'] == 1) & (counts['U22'] < 2),
    (counts['DP'] == 3) & (counts['U22'] >= 2),
    (counts['DP'] == 3)
],
[
    "YDP + max U22",
    "YDP + non-max U22",
    "2 overage DP + <MaxTAM overage DP + max U22",
    "3 overage DP + 1 allowed U22"
],default = "No 3rd DP")
counts
```

```python
counts = counts[['season','team_id','category']]
final_gplus_df = pd.merge(team_roster_gplus, counts)
final_gplus_df
```
