```python
from scipy.stats import shapiro
shapiro(data)
```

```python
def why_cant_you_just_be_normal(position, action_type):
    slice_gplus = composite[(composite.general_position == position) & (composite["data.action_type"] == action_type)]
    if (len(slice_gplus) == 0):
#         print(f"no data for Combo of {position} / {action_type}")
        return

    data = slice_gplus["data.goals_added_raw"]
    result = shapiro(data)
    if result.pvalue >= 0.5:
        print(f"Combo of {position} / {action_type} is normal with p-value {result.pvalue}")
    else:
        print(f"Combo of {position} / {action_type} is NON normal with p-value {result.pvalue}")
        
action_types = composite["data.action_type"].unique().tolist()
positions = composite["general_position"].unique().tolist()

for t in action_types:
    for p in positions:
        why_cant_you_just_be_normal(p, t)
```

```python
base_range = np.linspace(0.01, 1.00, 100)
base_range
```

```python
def percentiles(position, action_type, year):
    slice_gplus = composite[(composite.general_position == position) & (composite["data.action_type"] == action_type) & (composite.season == year)]
    if (len(slice_gplus) == 0):
        #print(f"no data for Combo of {position} / {action_type}")
        return pd.DataFrame()

    data = slice_gplus["data.goals_added_raw"] * 96 / slice_gplus["minutes_played"]
    return pd.DataFrame({ "position" : position, "action_type" : action_type, "season" : year, "pct" : base_range, "value" : data.quantile(base_range)})


years = composite["season"].unique().tolist()
percentile_composite = pd.DataFrame()
for t in action_types:
    for p in positions:
        for y in years:
            df = percentiles(p, t, y)
            percentile_composite = percentile_composite.append(df, ignore_index=True)
        
percentile_composite
```

```python
import pandas as pd
import json

seasons = range(2013, 2023)

gplus_data = pd.DataFrame()
for yr in seasons:
    print(f"Grabbing {yr}")
    url = f"https://app.americansocceranalysis.com/api/v1/mls/players/goals-added?season_name={yr}&split_by_teams=true&split_by_seasons=true"
    tmp = pd.read_json(url)
    
    print(f"Grabbing GK {yr}")
    gk_url = f"https://app.americansocceranalysis.com/api/v1/mls/goalkeepers/goals-added?season_name={yr}&split_by_teams=true&split_by_seasons=true"
    tmp_gk = pd.read_json(gk_url)
    tmp_gk['general_position'] = 'GK'
    tmp_gk
    
    gplus_data = gplus_data.append(tmp, ignore_index=True)
    gplus_data = gplus_data.append(tmp_gk, ignore_index=True)
    
gplus_data
```

```python
json_gk_expl_txt = json.loads(gplus_data.explode('data').to_json(orient="records"))
gplus_expl_flat = pd.json_normalize(json_gk_expl_txt)
gplus_expl_flat.to_csv('./season-g+.csv', index=False)
gplus_expl_flat.head()
```

```python
import numpy as np
def percentiles(position, action_type, year):
    slice_gplus = gplus_expl_flat[(gplus_expl_flat.general_position == position) & (gplus_expl_flat["data.action_type"] == action_type) & (gplus_expl_flat.season_name == year)]
    if (len(slice_gplus) == 0):
        #print(f"no data for Combo of {position} / {action_type}")
        return pd.DataFrame()

    data = slice_gplus["data.goals_added_raw"]
    adj_data = data * 96 / slice_gplus["minutes_played"]
    return pd.DataFrame({ "position" : position, "action_type" : action_type, "season" : year, "pct" : base_range, "p96" : adj_data.quantile(base_range), "pSzn" : data.quantile(base_range)})

base_range = np.linspace(0.01, 1.00, 100)
years = gplus_expl_flat["season_name"].unique().tolist()
action_types = gplus_expl_flat["data.action_type"].unique().tolist()
positions = gplus_expl_flat["general_position"].unique().tolist()
percentile_composite = pd.DataFrame()
for t in action_types:
    for p in positions:
        for y in years:
            df = percentiles(p, t, y)
            percentile_composite = percentile_composite.append(df, ignore_index=True)
        
percentile_composite
```

```python
# roster g+ breakdowns
grouped_gplus = gplus_expl_flat.groupby(['season_name','team_id','player_id','general_position']).agg({
    'data.goals_added_raw': ['sum'], 
    'minutes_played' : ['mean']
}).reset_index()

grouped_gplus.columns = grouped_gplus.columns.droplevel(level=1)
grouped_gplus
```

```python
grouped_gplus['total'] = grouped_gplus['data.goals_added_raw']
grouped_gplus['p96'] = grouped_gplus['data.goals_added_raw'] * 96 / grouped_gplus["minutes_played"]
grouped_gplus = grouped_gplus[['season_name', 'team_id', 'player_id','general_position', 'minutes_played','total', 'p96']]

grouped_gplus
```

```python

team_breakdown_gplus = grouped_gplus.groupby(['season_name','team_id','general_position']).apply(
        lambda x: pd.Series([
            np.mean(x['total']),
            np.mean(x['p96']),
            np.average(x['p96'], weights=x['minutes_played'])
        ], index=['total_avg', 'p96_avg', 'p96_weighted_avg'])
    ).reset_index()

team_breakdown_gplus
```

```python
team_breakdown_gplus['total_avg_rank'] = team_breakdown_gplus.groupby(['season_name','general_position'])['total_avg'].rank(ascending=False)
team_breakdown_gplus['p96_avg_rank'] = team_breakdown_gplus.groupby(['season_name','general_position'])['p96_avg'].rank(ascending=False)
team_breakdown_gplus['p96_weighted_avg_rank'] = team_breakdown_gplus.groupby(['season_name','general_position'])['p96_weighted_avg'].rank(ascending=False)

team_breakdown_gplus.to_csv('./team_position_breakdown.csv', index=False)
team_breakdown_gplus
```
